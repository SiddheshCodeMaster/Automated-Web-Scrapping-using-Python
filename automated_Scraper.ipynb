```python
from autoscraper import AutoScraper
```

```python
amazon_url = "https://www.amazon.in/s?k=iphones"

wanted_list = ["₹54,490","Apple iPhone 12 (128GB) - Blue"]
```

```python
# Building an autoscraper:

scraper = AutoScraper()
result = scraper.build(amazon_url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(amazon_url, grouped=True)
```

```python
scraper.set_rule_aliases({'rule_pl0p':'Title','rule_skri':'Price'})
scraper.keep_rules(['rule_pl0p','rule_skri'])
scraper.save('amazon-search')
```

```python
results=scraper.get_result_similar('https://www.amazon.in/s?k=mi+phones+under+15000',group_by_alias=True)
```

```python
results['Price']
```

```python
import pandas as pd

df1 = pd.DataFrame(results['Title'])
df2 = pd.DataFrame(results['Price'])

dfTotal = pd.concat([df1,df2], axis = 1, join='inner')
dfTotal.head()
```
